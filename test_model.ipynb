```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig


tokenizer = AutoTokenizer.from_pretrained("LuuNgoc2k2/Law-Llama-v1", add_eos_token=True, padding_side='right')
model = AutoModelForCausalLM.from_pretrained(   
    'LuuNgoc2k2/Law-Llama-v1',
    torch_dtype=torch.bfloat16,
    #quantization_config=bnb_config, # If you need 
    device_map="auto",
    use_cache=True,
)

tokenizer.pad_token = tokenizer.eos_token
```

```python
PROMPT = """
Nếu bạn không biết câu trả lời cho một câu hỏi, hãy trẳ lời là bạn không biết và vui lòng không chia sẻ thông tin sai lệch.
Dựa trên khẳng định: {in} Hãy trả lời câu hỏi: {input}
### Trả lời:
"""
passage = """
1. Người nào mua bán trái phép chất độc nguy hiểm, phóng xạ, chất cháy nguy hiểm, chất nổ, hàng giả có khối lượng từ 100 kilôgam đến dưới 500 kilôgam hoặc thu lợi bất hợp pháp từ 50.000.000 đồng đến dưới 300.000.000 đồng hoặc đã bị xử phạt vi phạm hành chính nhưng chưa hết thời hạn được coi là chưa bị kỷ luật hoặc đã bị kết án mà chưa được xóa án tích về tội này hoặc về một trong các tội quy định tại các điều 168, 169, 170, 171, 175 và 290 của Bộ luật này thì bị phạt cải tạo không giam giữ đến 03 năm hoặc phạt tù từ 06 tháng đến 03 năm.
2. Phạm tội thuộc một trong các trường hợp sau đây, thì bị phạt tù từ 02 năm đến 07 năm:
a) Có tổ chức;
b) Lợi dụng chức vụ, quyền hạn;
c) Mua bán về số lượng lớn;
d) Vận chuyển qua biên giới;
đ) Tái phạm nguy hiểm.
3. Phạm tội thuộc một trong các tội quy định ở khoản 2 Điều này do hoàn cảnh đặc biệt khó khăn mà chỉ có tác động duy nhất của việc áp dụng biện pháp tư pháp mới gỡ ra được hoàn cảnh đó, thì bị phạt tiền từ 10.000.000 đồng đến 100.000.000 đồng hoặc phạt cải tạo không giam giữ từ 01 năm đến 02 năm.
4. Người phạm tội còn có thể bị phạt tiền từ 5.000.000 đồng đến 50.000.000 đồng, cấm đảm nhiệm chức vụ, cấm hành nghề kinh doanh ngành, nghề liên quan từ 01 năm đến 05 năm."
Như vậy, nếu mua bán trái phép chất ma tuy có khối lượng từ 100kg đến dưới 500kg hoặc thu lợi bất hợp pháp trên 300 triệu đồng thì sẽ bị phạt tù từ 6 tháng đến 7 năm.
Lưu ý: Mức hình phạt nêu trên là mức phạt đối với người phạm tội tự thú, đầu thú. Nếu người phạm tội không tự thú, không đầu thú thì mức phạt sẽ tăng thêm từ 10 % đến 20 %.
Ngoài ra, người phạm tội còn có thể bị truy cứu trách nhiệm hình sự khác theo quy định của Bộ luật Hình sự 2015.
"""

question = """Người buôn bán trái phép ma túy thì bị đi tù bao nhiêu năm?"""

text = PROMPT.format_map({
    'in': passage,
    'input': question,
})
input_ids = tokenizer(text, return_tensors='pt', add_special_tokens=False).to('cuda')
```

```python
generated_ids = model.generate(
    input_ids=input_ids['input_ids'],
    max_new_tokens=1024,
    do_sample=True,
    top_p=0.95,
    top_k=40,
    temperature=0.3,
    repetition_penalty=1.1,
    no_repeat_ngram_size=7,
    #num_beams=5,
 )

a = tokenizer.batch_decode(generated_ids)[0]
# print(a.split('### Trả lời:')[1])
print(a)
```

```python
import json
def load_json_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

test = load_json_file('/kaggle/input/law-qa1/Law-QA-test.json')
```

```python
def split_passage(passage, max_tokens=900):
    # Split the passage into chunks with approximately max_tokens per chunk
    chunks = []
    current_chunk = ""
    current_tokens = 0
    
    for word in passage.split():
        word_tokens = len(tokenizer.tokenize(word))
        if current_tokens + word_tokens <= max_tokens:
            current_chunk += " " + word
            current_tokens += word_tokens
        else:
            chunks.append(current_chunk.strip())
            current_chunk = word
            current_tokens = word_tokens
    
    if current_chunk:
        chunks.append(current_chunk.strip())
    
    return chunks
```

```python
def replace_unk_with_newline(text):
    # Replace <unk> with newline characters
    cleaned_text = text.replace("<unk>", "\n")
    return cleaned_text
```
